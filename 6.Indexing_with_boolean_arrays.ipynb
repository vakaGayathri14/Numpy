```python
arr1 = np.random.randint(low=1,high=100,size=20).reshape(4,5)
arr1
```

```python
(arr1>50).shape
```

```python
#indexing using boolean arrays / filtering
arr1[arr1>50]
```

```python
arr1[(arr1>50) & (arr1 %2 != 0)] #odd numbers
```

```python
arr1[(arr1>50) & (arr1 %2 != 0)] =0 #replaces the conditionsatisfied values to 0
```
