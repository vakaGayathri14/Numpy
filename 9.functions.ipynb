```python
import numpy as np
```

```python
np.random.random() # this gives me any random values from 0 to 1, every time i run this it will generate a new value
```

```python
# if i want to generate same random value all the time then we have to use the function called seed
np.random.seed(1)
np.random.random()
```

```python
# if we want to generate a random value from a given range then 
np.random.uniform(3,10) # here 3 is my starting number and 10 is my ending number, this number changes at each run 
```

```python
# if we want to generate a random value from a given range of particular length then
np.random.uniform(1,100,10) # here 10 as length of the list
```

```python
np.random.uniform(1,100,10).reshape(2,5) # 2d array
```

```python

# if i want to replace all the odd value = -1 then 
a = np.random.randint(1,10,6)
a
```

```python
a[a%2==1]=-1
a
```

```python
# if i want to create another array for the replaced values rather than modifying it in array for that we have to use where
# where(condition,True,False)

a = np.random.randint(1,20,6)
a
```

```python
np.where(a%2==1,-1,a) #where returns me anoher array
```

```python
a
```

```python
out = np.where(a%2==1,-1,a)
out
```
