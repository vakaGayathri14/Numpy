```python
import numpy as np
```

```python
#Ravel function : it will bring higher dimensional arrays to lower dimension

arr1 = np.array([[6,7],[8,9],[10,11]])
arr1
```

```python
# dimension of the above array is 
arr1.ndim # so 2 dimension
```

```python
arr1.ravel() # here ravel function bring any 3d 4d or 1d arrays in to 1 dimension without any data loss
```

```python
# transpose row->column, column -> row
arr1.transpose()
```

```python
#stacking -> we can combine 2 arrays
arr2 = np.arange(6).reshape(2,3)
arr2
```

```python
arr3 = np.arange(12,18).reshape(2,3)
arr3
```

```python
# inorder to do  stacking we take a matrix of same length
np.hstack((arr2,arr3)) # stacking horizontly
```

```python
np.vstack((arr2,arr3))  #vertical stacking
```

```python
# splitting
np.hsplit(arr2,3) # horizontal splitting
```

```python
np.vsplit(arr2,2)
```

```python
# fancy indexing
arr4 = np.arange(24).reshape(6,4)
arr4
```

```python
# 1st , 3rd and 5th row - you can use fancy indexing
arr4[[0,2,4]] # here i am getting all the rows using index numbers 
```
