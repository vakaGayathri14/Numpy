```python
import numpy as np
```

```python
#scenario 1
a1 = np.arange(8).reshape(2,4)
a2 = np.arange(8,16).reshape(2,4)

print(a1)
print(a2)
```

```python
a1 + a2
```

```python
#scenario 2
a3 = np.arange(9).reshape(3,3)
a4 = np.arange(3).reshape(1,3)
print(a3)
print(a4)
```

```python
a3 + a4 # broadcasting:- ability of numpy to perform airthmetical operations on disimilar arrays
```

```python
# if x=1 and y = n then also operation will takes place(Same Dimension) like 3 is common
a5 = np.arange(3).reshape(1,3)
a6 = np.arange(12).reshape(4,3)

print(a5)
print(a6)
```

```python
a5 + a6
```

```python
# if y = 1 and x=m then also operation will take place, even if they are not of the same dimension

a7 = np.arange(4).reshape(4,1)
a8 = np.arange(12).reshape(4,3)

print(a7)
print(a8)
```

```python
a7 + a8
```

```python
# if x =1 and y!=n then also operation will not take place
a9 = np.arange(3).reshape(1,3)
a10 = np.arange(16).reshape(4,4)

print(a9)
print(a10)
```

```python
a9 + a10 #operands could not be broadcast together with shapes (1,3) (4,4) 
```

```python
# if x =1 and n=m operation to take place

a11 = np.arange(3).reshape(1,3)
a12 = np.arange(3).reshape(3,1)

print(a11)
print(a12)
```

```python
a11 + a12
```

```python
# if x =1 and y = 1 then the operation will take place no matter what

a13 = np.arange(1).reshape(1,1)
a14 = np.arange(20).reshape(4,5)

print(a13)
print(a14)
```

```python
a13 + a14
```
